```python
from flask import Flask, request, jsonify
from flask_cors import CORS
from pyngrok import ngrok
import threading
import time
import json
from pathlib import Path
import torch
import sys

# ----------------------------------------
# IMPORTANT:
# tokenizer + model are ALREADY LOADED
# outside this file (as you said).
# I WILL NOT redefine them here.
# ----------------------------------------

app = Flask(__name__)
CORS(app)
PORT = 5000


# ======= CLEAN / EXTRACT FUNCTIONS ========= #

def clean_generated(text: str) -> str:
    """Remove tags + unwanted prefixes."""
    text = text.replace("<think>", "").replace("</think>", "")
    text = text.replace("Assistant:", "").replace("User:", "")
    return text.strip()


def extract_new_text(full_output, prompt_text):
    """Extract only the new assistant reply."""
    if prompt_text in full_output:
        new = full_output[len(prompt_text):].strip()
    else:
        new = full_output.strip()

    new = new.replace("Assistant:", "").replace("User:", "")
    return new.strip()


# ======= MEMORY + SYSTEM MESSAGE ========= #

messages = [
    {
        "role": "system",
        "content": (
           "I'm a helpful and friendly AI assistant that speaks only in English. "
            "I remember what the user tells you during this session and across sessions using stored memory. "
            "If the user tells you their name, remember it and use it naturally later. "
            "Be professional, detailed, and never use any random name like 'Charlie' or 'John' unless the user says so. "
        ),
    }
]


def prune_messages(tokenizer, messages, reserve_new_tokens=512):
    text = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    tokens = tokenizer.encode(text)
    max_len = getattr(tokenizer, "model_max_length", 4096)

    while len(tokens) + reserve_new_tokens > max_len and len(messages) > 2:
        messages.pop(1)
        text = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
        tokens = tokenizer.encode(text)

    return messages



# =================== CHAT ENDPOINT =================== #

@app.route("/chat", methods=["POST"])
def chat():
    try:
        data = request.get_json()
        user_message = data.get("message", "")

        if not user_message:
            return jsonify({"reply": "No message provided."}), 400

        # Add user message
        messages.append({"role": "user", "content": user_message})
        prune_messages(tokenizer, messages)

        # Prepare input for model
        text_input = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
        inputs = tokenizer(text_input, return_tensors="pt").to("cuda")

        # Generate
        outputs = model.generate(
            **inputs,
            max_new_tokens=300,
            temperature=0.25,
            top_p=0.9,
            do_sample=True,
            eos_token_id=tokenizer.eos_token_id,
            pad_token_id=tokenizer.eos_token_id,
        )

        generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

        # ------------------------------------------------
        # REMOVE REPEATED HISTORY â€” KEEP LAST ASSISTANT
        # ------------------------------------------------
        lower_generated = generated_text.lower()
        if "assistant" in lower_generated:
            parts = lower_generated.split("assistant")
            last_part = parts[-1].strip()

            idx = generated_text.lower().rfind(last_part)
            if idx != -1:
                generated_text = generated_text[idx:].strip()
            else:
                generated_text = last_part

        # Extract new reply
        ai_response = extract_new_text(generated_text, text_input)
        ai_response = clean_generated(ai_response)

        # Save assistant reply
        messages.append({"role": "assistant", "content": ai_response})

        return jsonify({"reply": ai_response})

    except Exception as e:
        print(f"Internal Server Error: {e}")
        return jsonify({"reply": f"âš ï¸ Backend Error: {str(e)}"}), 500



# ================== SERVER LAUNCH ================== #

if __name__ == "__main__":

    def run_flask():
        print("Starting Flask server...")
        app.run(host="0.0.0.0", port=PORT, debug=False, use_reloader=False)

    flask_thread = threading.Thread(target=run_flask)
    flask_thread.start()

    print("Waiting for server start (5 seconds)...")
    time.sleep(5)

    try:
        ngrok.kill()
        public_url = ngrok.connect(PORT, proto="http").public_url

        print("\n" + "=" * 50)
        print(f" Work on port: {PORT}")
        print(f" Ngrok link: {public_url}")
        print(f" Use the link with/chat: {public_url}/chat")
        print("=" * 50 + "\n")

    except Exception as e:
        print(f"Ngrok Error: {e}")
        print("Auth Token.")


    try:
        while True:
            time.sleep(1)
    except KeyboardInterrupt:
        print("\nStopping server...")
        ngrok.kill()
```
